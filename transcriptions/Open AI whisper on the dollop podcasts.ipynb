```python
from youtubesearchpython import *
import subprocess
import os
import whisper
```

```python
# set path

path = "C:/Users/Declan/Documents/DataScienceProjects/dollop"
```

```python
# find all videos

playlist = Playlist('https://www.youtube.com/playlist?list=PL5JtZ5eCov8KMT_5c8mjP4DfiSMR-pWvd') #the dollop playlist


print(f'Videos Retrieved: {len(playlist.videos)}')

while playlist.hasMoreVideos:
    print('Getting more videos...')
    playlist.getNextVideos()
    print(f'Videos Retrieved: {len(playlist.videos)}')

print('Found all the videos.')
```

```python
# Download each video

for video in playlist.videos:
    
    if any(video['title'] in s for s in os.listdir(path)):
        pass
    
    else:
        print('downloading ' + video['title'])
        url = video['link']
        url = url[:url.find('&list')]

        command = "yt-dlp -P" + path + "-x --audio-format mp3" + " " + url
        subprocess.call(command, shell=True)
```

```python
# Clean up files in directory
for file in os.listdir(path):
    try:
        file_new = file[file.find('#'):]

        os.rename(path+'/'+file,path+'/'+file_new)
    except:
        pass
```

```python
# Create directory for whisper files
if os.path.exists(path+"/transcriptions"):
    pass
else:
    os.mkdir(path+"/transcriptions")

```

```python
model = whisper.load_model("small")

transcribed_files = os.listdir(path+"/transcriptions")


for file in os.listdir(path):
    if file =='transcriptions':
        continue
        
    temp_file = file.replace('.mp3',"")
    temp_file = temp_file+".txt"

    if temp_file in transcribed_files:
        print('passing', temp_file)
        pass
    else:
        print(file)
        
        result = model.transcribe(path + '/' + file)
        
        with open(path+"/transcriptions/" + file.replace(".mp3","") + ".txt", "w") as f:
            
            result = result["text"].encode('ascii','ignore')
            result = result.decode()
                        
            f.write(result)
```
